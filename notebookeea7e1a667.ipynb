```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler,LabelEncoder

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Load the dataset
df=pd.read_csv("/kaggle/input/heart-failure-prediction/heart.csv")
df.head(7)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Visualization
y=df['HeartDisease']
plot_sb = sns.countplot(df,x=y, label='Total')
Rain, NotRain =y.value_counts()
print('Have Heart Disease: ',Rain)
print('Not Have Heart Disease : ',NotRain)
```

```python
df.hist(figsize=(12, 10))
plt.suptitle('Histogram for number of categoricals')
plt.show()
```

```python
# Preprocessing
sex=LabelEncoder()
cpt=LabelEncoder()
recg=LabelEncoder()
ea=LabelEncoder()
st=LabelEncoder()
df['Sex']=sex.fit_transform(df['Sex'])
df['ChestPainType']=cpt.fit_transform(df['ChestPainType'])
df['RestingECG']=recg.fit_transform(df['RestingECG'])
df['ExerciseAngina']=ea.fit_transform(df['ExerciseAngina'])
df['ST_Slope']=st.fit_transform(df['ST_Slope'])
```

```python
df.head()
```

```python
corr_matrix=df.corr(method='pearson')
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
```

```python
thresh_hold=0.2
select_feat=corr_matrix.index[abs(corr_matrix['HeartDisease'])>=thresh_hold].to_list()
select_feat.remove('HeartDisease')
print(select_feat)
```

```python
selected_features=df[select_feat]
selected_features.head()
```

```python
target=df['HeartDisease']
target
```

```python
scaler=StandardScaler()
selected_features=scaler.fit_transform(selected_features)
selected_features
```

```python
# Splitting the Data
x_train,x_test,y_train,y_test=train_test_split(selected_features,target,test_size=0.2)
```

```python
# Build SVM Model
from sklearn.svm import SVC

SVM_model=SVC()
SVM_model.fit(x_train,y_train)
```

```python
svm_pred=SVM_model.predict(x_test)
```

```python
# Evaluate SVM Model
accuracy=accuracy_score(y_test,svm_pred)

print("Accuracy=",int(accuracy*100),'%')  
```

```python
import joblib

# Save the model to a file
joblib.dump(SVM_model, 'svm_model.pkl')
```

```python
# Example Input Parameters for "No Heart Disease"
custom_input = [[35, 1, 1, 0, 1, 160, 0, 0.0, 2]]  # Replace with actual input values

# Scale the input using the fitted StandardScaler
custom_input_scaled = scaler.transform(custom_input)

# Predict using the trained SVM model
prediction = SVM_model.predict(custom_input_scaled)

# Display Prediction Result
if prediction[0] == 0:
    print("Prediction: No Heart Disease")
else:
    print("Prediction: Heart Disease")
```
